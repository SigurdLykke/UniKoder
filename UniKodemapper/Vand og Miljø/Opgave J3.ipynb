```python
from sympy import *
import pandas as pd

M_co2=44.1 #g/mol
M_caco3=100.0869 #g/mol

data = {
    'Jordtype': ['Muld', 'Mor', 'Lavmosetørv', 'Højmosetørv'],
    'Vægt af tør jord inden afbrænding (g)': [10.0, 5.0, 10.0, 5.0],
    'Vægt af CO2 udviklet ved afbrænding (g)': [0.238, 8.130, 6.986, 9.578],
    'Vægt af CO2 udviklet ved tilsætning af syre (g)': [0.2, 0, 1.0, 0],
    'Kvælstofvægt %': [0.145, 1.059, 0.789, 1.161],
}

df = pd.DataFrame(data)

# Funktion til at beregne uorganisk kulstof (vægt % CaCO3)
def beregn_uc(row):
    co2_afbraending = row['Vægt af CO2 udviklet ved afbrænding (g)']
    co2_syre = row['Vægt af CO2 udviklet ved tilsætning af syre (g)']
    tør_jord = row['Vægt af tør jord inden afbrænding (g)']
    # Beregning af CaCO3 vægt% baseret på CO2 mængder, antagelsen af forholdet mellem caco3 og co2 er 1:1 og tør jord vægt
    vægt_procent_caco3 = (co2_syre/M_co2)*M_caco3 * 100 / tør_jord
    return vægt_procent_caco3

# Tilføje kolonnen 'vægt% CaCO3 (1)' til DataFrame
df['vægt% CaCO3 (1)'] = df.apply(beregn_uc, axis=1)

# Funktion til at beregne organisk kulstof baseret på kvælstofvægt%
def beregn_organisk_stof(row):
    organisk_carbon_procent = row['vægt% CaCO3 (1)']* 
    return organisk_carbon_procent

# Tilføje kolonnen 'Organisk carbon' til DataFrame
df['Organisk carbon (2)'] = df.apply(beregn_organisk_stof, axis=1)

df
```
