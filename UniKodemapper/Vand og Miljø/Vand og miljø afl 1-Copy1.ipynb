```python
import sympy as sym
import matplotlib.pyplot as plt
import math

# Definer konstanter
m_start=100/1000 #kg
m_tør=70/1000 #kg
V_prøve=57/(100**3) #m^3
densi_standard= 2650 #kg/m^3
densi_vand = 1000 #kg/m^3

# Beregning af volumenvægt
V_vægt= m_tør/V_prøve

# Beregning af vægt % vand
m_procentvand=(m_start-m_tør)*100/m_tør

# Beregning af porøsitet
porøsitet_jordprocent=(1-V_vægt/densi_standard)*100

# Beregning af volumenprocent vand
V_procentvand = (V_vægt/densi_vand)*m_procentvand

# Udskrivning af resultater
print(f"1.Volumenvægt: {V_vægt} kg/m³")
print(f"2.Vægt % vand: {m_procentvand} %")
print(f"3.Porøsitet af jorden: {porøsitet_jordprocent} %")
print(f"4.Jordens Volumetriske Vandindhold: {V_procentvand} %")
print(f"5. ")

# Lagkage diagram
mineraliske_partikler = 100-3-porøsitet_jordprocent
organisk_materiale = 3
vand = (V_procentvand/100)*(porøsitet_jordprocent/100)*100
luft = ((100-V_procentvand)/100)*(porøsitet_jordprocent/100)*100
# eller luft = 100-(100-3-porøsitet_jordprocent)-(3)-((V_procentvand/100)*(porøsitet_jordprocent/100)*100)

sizes = [mineraliske_partikler, organisk_materiale, vand, luft]
labels = ['Mineraliske partikler', 'Organisk materiale', 'Vand', 'Luft']
colors = ['brown', 'yellowgreen', 'darkblue', 'lightskyblue']
explode = (0.1, 0, 0, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
print(f"Mineraliske partikler: {mineraliske_partikler} %")
print(f"Organisk materiale: {organisk_materiale} %")
print(f"Vand: {vand} %")
print(f"Luft: {luft} %")
print(f" ")

# Beregning af mg nitrat per liter
h = 30 #cm
CM_3=57 #cm^3
M_3=CM_3/100**3 #m^3
CM_2=CM_3/h #cm^2
M_2=CM_2/100**3 #m^2
L=M_3*1000 #L
M_nitrat=62 #g/mol
M_nitrogen=14.01 #g/mol
c_nitrat=0.015 #mol/L
m_nitrat=M_nitrat*c_nitrat*L #g
m_nitrogen=M_nitrogen*c*L #g
mg_nitrat=m_nitrat*1000
mg_nitrogen=m_nitrogen*1000
res6=(mg_nitrat/L)
res7=(mg_nitrogen/M_2)*(V_procentvand/100)

# Udskrivning af resultater for nitrat
print(f"6. {res6} mg/L")
print(f"7. {res7} mg/m^2")
```
