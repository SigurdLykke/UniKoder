```python
from sympy import *
import pandas as pd

data = {
    'Horisont navn': ['Ap', 'Bw', 'BC', 'C'],
    'Tykkelse': [27, 31, 33, 69],
    'Finsand': [16, 20, 18, 13],
    'Silt': [20, 24, 19, 16],
    'Ler': [9, 8, 6, 26],
    'Organisk stof': [2.4, 0.6, 0.2, 0.2]
}

df = pd.DataFrame(data)

# Funktion til at beregne PTV
def beregn_ptv(org_stof, ler, silt, finsand):
    return (1.96 * org_stof) + (0.02 * ler) + (0.34 * silt) + (0.17 * finsand) + 2.26

# Anvende funktionen til hver r√¶kke i DataFrame
df['PTV (vol. %)'] = df.apply(lambda row: beregn_ptv(row['Organisk stof'], row['Ler'], row['Silt'], row['Finsand']), axis=1)

df

def beregn_rzk (tykkelse, beregn_ptv):
    return (beregn_ptv*tykkelse)/10
df['RZK (mm vand)']=df.apply(lambda row: beregn_rzk(row['PTV (vol. %)'], row['Tykkelse']), axis=1)

df

```
